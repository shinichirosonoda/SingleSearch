```python
fig = plt.figure(figsize=(15,90))
[rows, cols] = [18,2]
ax1 = fig.add_subplot(rows, cols, 1)
ax2 = fig.add_subplot(rows, cols, 2)
ax3 = fig.add_subplot(rows, cols, 3)
ax4 = fig.add_subplot(rows, cols, 4)
ax5 = fig.add_subplot(rows, cols, 5)
ax6 = fig.add_subplot(rows, cols, 6)
ax7 = fig.add_subplot(rows, cols, 7)
ax9 = fig.add_subplot(rows, cols, 9)
ax10 = fig.add_subplot(rows, cols, 10)
ax11 = fig.add_subplot(rows, cols, 11)
ax12 = fig.add_subplot(rows, cols, 12)
ax13 = fig.add_subplot(rows, cols, 13)
ax14 = fig.add_subplot(rows, cols, 14)


# folder = "./single_peak_search_data"
# folder = "./single_peak_search_data/FOVDevice"
# folder = "./single_peak_search_data/Alumium_1"
folder = "./single_peak_search_data/WP1"
files = glob.glob(folder + "/*.txt")


slow_q_list = []
fast_q_list = []
slow_pfv_q_list = []
fast_pfv_q_list = []
slow_diff_q_list = []
fast_diff_q_list = []
freq_rate_list = []
freq_rate_diff_list = []
slow_res_freq_list = []
fast_res_freq_list = []
for _k, file in enumerate(files[:]):
# for file in files[2:3]:
# for file in files[-2:]:
    print(file, files[0][-13:-4])
    
    # Single_Search_Data -----------------------------
    f = open(file, 'r')
    data = f.read()
    f.close()

    slow_freq_list = [int(_i) for _i in data.split("slow_freq_list[bit]")[1].split("slow_phase_list[clk]")[0].split("[")[1].split("]")[0].split(",")]
    slow_phase_list = [float(_i) for _i in data.split("slow_phase_list[clk]")[1].split("slow_vpp_list[bit]")[0].split("[")[1].split("]")[0].split(",")]
    slow_vpp_list = [float(_i) for _i in data.split("slow_vpp_list[bit]")[1].split("temperature[deg]")[0].split("[")[1].split("]")[0].split(",")]
    
    fast_freq_list = [int(_i) for _i in data.split("slow_freq_list[bit]")[2].split("slow_phase_list[clk]")[0].split("[")[1].split("]")[0].split(",")]
    fast_phase_list = [float(_i) for _i in data.split("slow_phase_list[clk]")[2].split("slow_vpp_list[bit]")[0].split("[")[1].split("]")[0].split(",")]
    fast_vpp_list = [float(_i) for _i in data.split("slow_vpp_list[bit]")[2].split("temperature[deg]")[0].split("[")[1].split("]")[0].split(",")]


#     # Aluminium_1
#     if _k==0:
#         slow_vpp_list = np.where(np.array(slow_vpp_list)>4200,0,np.array(slow_vpp_list))
#     if _k==2:
#         slow_vpp_list = np.where(np.array(slow_vpp_list)>4220,0,np.array(slow_vpp_list))
#     if _k==8:
#         slow_vpp_list = np.where(np.array(slow_vpp_list)>4173,0,np.array(slow_vpp_list))

#     # WP11
#     if _k==1:
#         slow_vpp_list = np.where(np.array(slow_vpp_list)>4100,4000,np.array(slow_vpp_list))

    # Search_Res_Cnd ----------------------------------
    slow_res_index = np.argmax(slow_vpp_list)
    fast_res_index = np.argmax(fast_vpp_list)
    
    
    # try_code ----------------------------------
    print("slow_res_phase: ", slow_phase_list[slow_res_index], "fast_res_phase: ", fast_phase_list[fast_res_index])

    
    ## Phase_Value _ NG ---------------------------------- 
#     a = slow_freq_list[slow_res_index]
#     b = slow_freq_list[np.where(np.array(slow_phase_list) > slow_phase_list[slow_res_index]+1000)[0][-1]]
#     slow_q_list.append(a/(a-b))
#     c = fast_freq_list[fast_res_index]
#     d = fast_freq_list[np.where(np.array(fast_phase_list) > fast_phase_list[fast_res_index]+500)[0][-1]]
#     fast_q_list.append(c/(c-d))
    
    
#     ## Phase_Fix_Value _ NG ---------------------------------- 
#     pfv_s_a = slow_freq_list[np.where(np.array(slow_phase_list) > 7680+0)[0][-1]]
#     pfv_s_b = slow_freq_list[np.where(np.array(slow_phase_list) > 7680+1100)[0][-1]]
#     slow_pfv_q_list.append(pfv_s_a/(pfv_s_a-pfv_s_b))
#     pfv_f_c = fast_freq_list[np.where(np.array(fast_phase_list) > 3670+0)[0][-1]]
#     pfv_f_d = fast_freq_list[np.where(np.array(fast_phase_list) > 3670+800)[0][-1]]
#     fast_pfv_q_list.append(pfv_f_c/(pfv_f_c-pfv_f_d))
    
    
    ## Vpp_Value ---------------------------------- 
    a = slow_freq_list[slow_res_index]
    b = slow_freq_list[np.where(np.array(slow_vpp_list[0:100]) < slow_vpp_list[slow_res_index]/np.sqrt(2))[0][-1]]
    slow_q_list.append(a/(a-b))
    c = fast_freq_list[fast_res_index]
    d = fast_freq_list[np.where(np.array(fast_vpp_list[0:100]) < fast_vpp_list[fast_res_index]/np.sqrt(2))[0][-1]]
    fast_q_list.append(c/(c-d))
    
    
    
    ## Phase_Diff_Value ---------------------------------- 
    slow_phase_diff_list = [(slow_phase_list[_j+1]-slow_phase_list[_j]) for _j in range(1,len(slow_phase_list)-1,1)]
    fast_phase_diff_list = [(fast_phase_list[_j+1]-fast_phase_list[_j]) for _j in range(1,len(fast_phase_list)-1,1)]
    slow_diff_res_index = np.argmin(slow_phase_diff_list)
    fast_diff_res_index = np.argmin(fast_phase_diff_list)    
    pdv_s_a = slow_freq_list[slow_diff_res_index]
#     pdv_s_b = slow_freq_list[np.where(np.array(slow_phase_diff_list[0:100]) > slow_phase_diff_list[slow_diff_res_index]/np.sqrt(2))[0][-1]]
    pdv_s_b = slow_freq_list[np.where(np.array(slow_phase_diff_list[0:100]) > -20)[0][-1]]
    slow_diff_q_list.append(pdv_s_a/(pdv_s_a-pdv_s_b))
    pdv_f_c = fast_freq_list[fast_diff_res_index]
#     pdv_f_d = fast_freq_list[np.where(np.array(fast_phase_diff_list[0:100]) > fast_phase_diff_list[fast_diff_res_index]/np.sqrt(2))[0][-1]]
    pdv_f_d = fast_freq_list[np.where(np.array(fast_phase_diff_list[0:100]) > -40)[0][-1]]
    fast_diff_q_list.append(pdv_f_c/(pdv_f_c-pdv_f_d))
    
    print("Phase_Diff_Value:", pdv_s_a, pdv_f_c)

    
    ## Freq_Rate_Value ----------------------------------     
    slow_res_freq_list.append(slow_freq_list[slow_res_index]/2**64*160000000)
    fast_res_freq_list.append(fast_freq_list[fast_res_index]/2**64*160000000)
    freq_rate_list.append(slow_freq_list[slow_res_index]/fast_freq_list[fast_res_index])
    freq_rate_diff_list.append(slow_freq_list[slow_diff_res_index]/fast_freq_list[fast_diff_res_index])

    
    # Figure ----------------------------------
    ax1.plot(slow_freq_list, slow_phase_list)
    ax1.scatter(slow_freq_list[slow_res_index], slow_phase_list[slow_res_index], color="red")
    ax2.plot(fast_freq_list, fast_phase_list)
    ax2.scatter(fast_freq_list[fast_res_index], fast_phase_list[fast_res_index], color="red")
    ax3.plot(slow_freq_list, slow_vpp_list)
    ax3.scatter(slow_freq_list[slow_res_index], slow_vpp_list[slow_res_index], color="red")
    ax4.plot(fast_freq_list, fast_vpp_list)
    ax4.scatter(fast_freq_list[fast_res_index], fast_vpp_list[fast_res_index], color="red")
    ax9.plot(slow_freq_list[1:-1], slow_phase_diff_list)
    ax9.scatter(slow_freq_list[slow_res_index], slow_phase_diff_list[slow_res_index], color="red")
    ax10.plot(fast_freq_list[1:-1], fast_phase_diff_list)
    ax10.scatter(fast_freq_list[fast_res_index], fast_phase_diff_list[fast_res_index], color="red")
#     ax9.scatter(slow_freq_list[slow_diff_res_index], slow_phase_diff_list[slow_diff_res_index], color="blue")
#     ax10.scatter(fast_freq_list[fast_diff_res_index], fast_phase_diff_list[fast_diff_res_index], color="blue")
    
#     ax1.scatter(slow_freq_list[slow_diff_res_index], slow_phase_list[slow_diff_res_index], color="blue")
#     ax2.scatter(fast_freq_list[fast_diff_res_index], fast_phase_list[fast_diff_res_index], color="blue")
#     ax3.scatter(slow_freq_list[slow_diff_res_index], slow_vpp_list[slow_diff_res_index], color="blue")
#     ax4.scatter(fast_freq_list[fast_diff_res_index], fast_vpp_list[fast_diff_res_index], color="blue")
    ###  
    ### ax7-blue vs red

    
print(slow_q_list)
print(fast_q_list)
ax5.scatter(np.arange(0,len(slow_q_list),1), slow_q_list)
ax6.scatter(np.arange(0,len(fast_q_list),1), fast_q_list)
ax7.scatter(np.arange(0,len(freq_rate_list),1), freq_rate_list, color="red")
# ax7.scatter(np.arange(0,len(freq_rate_diff_list),1), freq_rate_diff_list, color="blue")
ax11.scatter(np.arange(0,len(slow_diff_q_list),1), slow_diff_q_list)
ax12.scatter(np.arange(0,len(fast_diff_q_list),1), fast_diff_q_list)
# ax13.scatter(np.arange(0,len(slow_pfv_q_list),1), slow_pfv_q_list)
# ax14.scatter(np.arange(0,len(fast_pfv_q_list),1), fast_pfv_q_list)

ax13.scatter(np.arange(0,len(slow_res_freq_list),1), slow_res_freq_list)
ax14.scatter(np.arange(0,len(fast_res_freq_list),1), fast_res_freq_list)



ax1.grid()
ax2.grid()
ax3.grid()
ax4.grid()    
ax5.grid()
ax6.grid() 
ax7.grid() 
ax9.grid() 
ax10.grid() 
ax11.grid() 
ax12.grid() 
ax13.grid() 
ax14.grid() 
```
